```python
import cv2
from matplotlib import pyplot as plt
import os
import pandas as pd
```

```python
# Please provide/change the data_dir path
# that expects to have the following folder:
# E-044_N10, E-044_N12, E-046_N10, E-046_N12, E-048_N10, E-048_N12, E084_N12, E-014_N38
# and ConeQuest_data.csv file

data_dir = "data"
```

```python
data_df = pd.read_csv(os.path.join(data_dir, "ConeQuest_data.csv"))

# Filter dataframe to load only positive samples (samples that have at least 1 cone)
cone_df = data_df.loc[(data_df["Number of Cones"]!=0)]

# Filter dataframe based on region
isidis_planitia_df = cone_df.loc[(cone_df["Region"]=="Isidis Planitia")].reset_index(drop=True)
acidalia_planitia_df = cone_df.loc[(cone_df["Region"]=="Acidalia Planitia")].reset_index(drop=True)
hypanis_df = cone_df.loc[(cone_df["Region"]=="Hypanis")].reset_index(drop=True)
```

```python
# Visualization for Isidis Planitia region
for index, row in isidis_planitia_df.iterrows():

  current_file = row["Patch Id"]
  print(current_file)

  folder_name = current_file.split("_")[0] + "_" + current_file.split("_")[1]
  input_path = os.path.join(data_dir, folder_name, "input_dir", current_file)
  mask_path = os.path.join(data_dir, folder_name, "output_dir", current_file)

  image = cv2.imread(input_path, 0)
  label = cv2.imread(mask_path, 0)

  plt.figure(figsize=(20, 12))
  plt.subplot(241)
  plt.xticks([])
  plt.yticks([])
  plt.title("Input")
  plt.imshow(image, cmap='gray')
  plt.subplot(242)
  plt.xticks([])
  plt.title("Ground Truth Mask")
  plt.yticks([])
  plt.imshow(label, cmap='gray')
  plt.show()

  if index == 5:
    break
```
